```python
ds['Date_of_Journey'] = pd.to_datetime(ds['Date_of_Journey'],infer_datetime_format=True)
```

```python
ds.info()
```

```python
ds['Day'] = ds['Date_of_Journey'].dt.day
ds['Month'] = ds['Date_of_Journey'].dt.month
ds['Year'] = ds['Date_of_Journey'].dt.year
ds.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
ds
```

```python
ds['Dep_Time'].unique()
```

```python
# ds['Dep_Time'].str.split(':')
ds['Dep_Hr'] = ds['Dep_Time'].str.split(':').str[0].astype(int)
ds['Dep_Se'] = ds['Dep_Time'].str.split(':').str[1].astype(int)
ds.drop('Dep_Time',axis=1,inplace=True)
```

```python
ds
```

```python
ds['Arrival_Time'].unique()
```

```python
ds['Arr_Hr'] = ds['Arrival_Time'].apply(lambda x: x.split(' ')[0].split(':')[0]).astype(int)
ds['Arr_Min'] = ds['Arrival_Time'].apply(lambda x: x.split(' ')[0].split(':')[1]).astype(int)
ds.drop('Arrival_Time',axis=1,inplace=True)
```

```python
ds
ds.info()
```

```python
ds['Total_Stops'].unique()
```

```python
ds['Total_Stops'].value_counts()
```

```python
ds['Total_Stops'] = ds['Total_Stops'].map({'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'non-stop':0,'nan':1})
```
