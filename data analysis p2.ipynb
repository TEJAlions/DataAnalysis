```python
data = pd.read_csv('https://raw.githubusercontent.com/sunnysavita10/Statistics-With-Python-TheCompleteGuide/main/clead_google_playstore_data')
data
```

```python
data = data.drop_duplicates(subset='App', keep='first')
```

```python
numData = [i for i in data.columns if data[i].dtype !='O' ]
catData = [i for i in data.columns if data[i].dtype =='O' ]
```
