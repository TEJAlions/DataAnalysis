```python
def changeData(val, size=False, price=False, install=False):
  if install:
    if val.replace('+','').replace(',','').isdigit():
      return int(val.replace('+','').replace(',',''))
    else: return np.NaN
  if price:
    if val.replace('$','').replace('.','').isdigit():
      return float(val.replace('$',''))
    else: return np.NaN
  if val.isnumeric():
    return int(val)
  elif 'M' in val:
    if val.replace('M','').isdigit() or val.replace('M','').replace('.','').isdigit():
      val = float(val.replace('M', ''))
      if(size): return int(val*1024)
      else: return int(val*1000000)
  elif 'K' in val:
    if val.replace('K','').isdigit() or val.replace('K','').replace('.','').isdigit():
      val = float(val.replace('K', ''))
      if(size): return float(val)
      else: return int(val*1000)
  elif size: return np.NaN
  return val
# changeSize('24.5K', True)
# changeData('$4.5', price=True)
# changeData('$4.5', False,True)
changeData('2000000+', install=True)
```

```python
#Chant gpt code

import pandas as pd

# URL of the CSV file
url = "https://raw.githubusercontent.com/sunnysavita10/Statistics-With-Python-TheCompleteGuide/main/googleplaystore.csv"

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(url)

# Display basic information about the dataset
# pd.set_option('display.max_columns', None)  # Show all columns
print(df.info())

# Step 1: Dealing with missing values and duplicates (same as before)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

# Step 2: Convert the 'Size', 'Price', 'Reviews', and 'Installs' columns to numeric (same as before)
df['Size'] = df['Size'].apply(lambda x: float(x.strip('M').replace('Varies with device', 'NaN')) if 'M' in x else float(x.strip('k').replace('Varies with device', 'NaN')) / 1000)
df['Price'] = df['Price'].apply(lambda x: float(x.strip('$')) if '$' in x else 0)
df['Reviews'] = df['Reviews'].astype(int)
df['Installs'] = df['Installs'].apply(lambda x: int(x.replace('+', '').replace(',', '')))

# Step 3: Handling categorical data (same as before)
category_dummies = pd.get_dummies(df['Category'], prefix='Category')
genre_dummies = pd.get_dummies(df['Genres'], prefix='Genre')

# Concatenate the one-hot encoded columns with the original DataFrame
df = pd.concat([df, category_dummies, genre_dummies], axis=1)

# Drop the original 'Category' and 'Genres' columns
df.drop(['Category', 'Genres'], axis=1, inplace=True)

# Step 4: Convert the 'Last Updated' column to datetime with error handling (same as before)
def convert_to_datetime(date_str):
    try:
        return pd.to_datetime(date_str)
    except:
        return None

df['Last Updated'] = df['Last Updated'].apply(convert_to_datetime)

# Final check after cleaning
print(df.info())
```
